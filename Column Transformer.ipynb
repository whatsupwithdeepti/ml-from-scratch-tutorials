COLUMN TRANSFORMER
instead of making different columns for ordinal and one hot encoding and merge them later , we can use this method to do this all steps simultaneously

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\ASUS\Downloads\covid_toy.csv")
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
transformer = ColumnTransformer(transformers=[('tnf1',SimpleImputer(),['fever']),
                                              ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
                                              ('tnf3',OneHotEncoder(sparse=False,drop='first'),['gender','city'])],remainder='passthrough') # this pass through will keep the column which doesnt require any option as it is + you pass the columns in tuples
# first you pass the new name of the column then the name of the operation you need to perform then the name of the column on which that operation is happening 
```

```python
transformer.fit_transform(X_train)
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test)
```
