```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Read the dataset
house_price_df=pd.read_excel(r"C:\Users\ASUS\OneDrive\Desktop\houseprices.xlsx")
house_price_df
```

```python
# Plotting the graph
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.scatter('Area','Price',data=house_price_df,color='red',marker='+')
plt.grid(True)
plt.show();
```

```python
# Initialize Linear Regression
reg=linear_model.LinearRegression()
```

```python
# Fit the model
reg.fit(house_price_df[['Area']],house_price_df.Price)
```

```python
# Predict the price based on area
reg.predict([[3300]])
```

```python
# coefficient
reg.coef_
```

```python
# intercept
reg.intercept_
```

```python
# Read the dataset to predict more prices
area=pd.read_excel(r"C:\Users\ASUS\OneDrive\Desktop\Area.xlsx")
area
```

```python
# prediction
predicted_df=reg.predict(area.values)
predicted_df
```

```python
# create a prices named column in the dataset and stored the predicted values in it
area['Price']=predicted_df
```

```python
# dataset after adding the column prices
area
```

```python
#Plotting Linear Regression 
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.scatter('Area','Price',data=house_price_df,color='red',marker='+')
plt.plot(house_price_df.Area,reg.predict(house_price_df[['Area']]),color='blue')
plt.grid()
plt.show();
```
