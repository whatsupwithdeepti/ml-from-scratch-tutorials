```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
titanic=pd.read_csv(r"C:\Users\ASUS\Downloads\train_and_test2.csv")
titanic.head()
```

1. Scatterplot ( Numerical - Numerical )

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

2. Bar Plot (Numerical - Categorial)

```python
plt.bar(flights['month'],flights['passengers'])
```

3. Boxplot (Numerical-Categorial)

```python
sns.boxplot(x=flights['month'],y=flights['passengers'])
```

4. Distplot(Numerical-Categorial)

```python
sns.distplot(flights['passengers'],hist=False)
```

5. Heatmap (Categorical-Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['2urvived']),annot=True)
```

```python
(titanic.groupby('Pclass').mean()['2urvived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Sex').mean()['2urvived']*100).plot(kind='bar')
```

6. Clustermap (Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titanic['sibsp'],titanic['2urvived']))
```

7. Pairplot

```python
sns.pairplot(iris,hue='species')
```

8. Lineplot (Numerical - Numerical)

```python
sns.lineplot(x=flights['year'],y=flights['passengers'])
```

9. Heatmap on flights dataset

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
